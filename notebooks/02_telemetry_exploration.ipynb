## 1. Setup & Imports

```python
import plotly.io as pio
pio.renderers.default = "notebook"

import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

from core.data.f1_loader import load_laps, load_session, load_telemetry

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: f'{x:.3f}')

print('✅ All imports successful')
```

```python
ver_tel = load_telemetry(2024, 'Bahrain', 'R', 'VER')
ham_tel = load_telemetry(2024, 'Bahrain', 'R', 'HAM')

print(f'VER telemetry shape: {ver_tel.shape}')
print(f'HAM telemetry shape: {ham_tel.shape}')
print(f'\nVER columns: {list(ver_tel.columns)}')
```

```python
ver_tel[[
    "Speed",
    "RPM",
    "Throttle",
    "Brake",
    "DRS",
    "nGear",
    "X",
    "Y",
    "Z",
    "Distance",
    "RelativeDistance",
    "SessionTime",
    "Time",
]].head()
```

```python
VER_COLOR = '#1E90FF'  # Red Bull blue
HAM_COLOR = '#00D2BE'  # Mercedes teal
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=ver_tel["Distance"],
    y=ver_tel["Speed"],
    mode='lines',
    name="VER",
    line=dict(width=2,color=VER_COLOR),
    hovertemplate=(
        f'<b>VER</b><br>'
        'Distance %{x}<br>'
        'Speed: %{y:.1f} km/h<br>'
        '<extra></extra>'
    )
))

fig.add_trace(go.Scatter(
    x=ham_tel["Distance"],
    y=ham_tel["Speed"],
    mode='lines',
    name="HAM",
    line=dict(width=2,color=HAM_COLOR),
    hovertemplate=(
        f'<b>HAM</b><br>'
        'Distance %{x}<br>'
        'Speed: %{y:.1f} km/h<br>'
        '<extra></extra>'
    )
))

fig.update_layout(
    title='2024 Bahrain GP — Speed Trace: VER vs HAM (Fastest Lap)',
    xaxis_title='Distance',
    yaxis_title='Speed',
    template='plotly_dark',
    hovermode='x unified',
    height=500,
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=ver_tel["Distance"],
    y=ver_tel["Throttle"],
    mode='lines',
    name="VER",
    line=dict(width=2,color=VER_COLOR),
    hovertemplate=(
        f'<b>VER</b><br>'
        'Distance %{x}<br>'
        'Throttle: %{y:.1f}<br>'
        '<extra></extra>'
    )
))

fig.add_trace(go.Scatter(
    x=ham_tel["Distance"],
    y=ham_tel["Throttle"],
    mode='lines',
    name="HAM",
    line=dict(width=2,color=HAM_COLOR),
    hovertemplate=(
        f'<b>HAM</b><br>'
        'Distance %{x}<br>'
        'Throttle: %{y:.1f}<br>'
        '<extra></extra>'
    )
))

fig.update_layout(
    title='2024 Bahrain GP — Throttle Trace: VER vs HAM (Fastest Lap)',
    xaxis_title='Distance',
    yaxis_title='Throttle',
    template='plotly_dark',
    hovermode='x unified',
    height=500,
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=ver_tel["X"],
    y=ver_tel["Y"],
    mode='lines',
    name="VER",
    line=dict(width=2,color=VER_COLOR),
    hovertemplate=(
        '<b>VER</b><br>'
        'X: %{x}<br>'
        'Y: %{y}<br>'
        '<extra></extra>'
    )
))

fig.add_trace(go.Scatter(
    x=ham_tel["X"],
    y=ham_tel["Y"],
    mode='lines',
    name="HAM",
    line=dict(width=2,color=HAM_COLOR),
    hovertemplate=(
        '<b>HAM</b><br>'
        'X: %{x}<br>'
        'Y: %{y}<br>'
        '<extra></extra>'
    )
))

fig.update_layout(
    title='2024 Bahrain GP — Racing Line: VER vs HAM (Fastest Lap)',
    xaxis_title='X',
    yaxis_title='Y',
    yaxis=dict(scaleanchor='x', scaleratio=1),
    template='plotly_dark',
    hovermode='x unified',
    height=500,
)

fig.show()
```

## 🔍 Key Findings — VER vs HAM Telemetry Analysis

| Channel | VER | HAM | Insight |
|---|---|---|---|
| Speed | Higher minimum corner speed throughout | Lower corner minimums | RB20 mechanical grip advantage allows later braking |
| Throttle | Earlier to 100% at corner exit | More hesitation before full throttle | Greater car confidence at exit — more downforce |
| Racing Line | Near identical to HAM | Near identical to VER | Both drivers know the optimal line — difference is execution not knowledge |

### Engineering Conclusion
The lap time delta between VER and HAM is not explained by straight line speed or racing line choice — both are nearly identical. The performance gap comes entirely from **corner entry speed** (later braking) and **corner exit throttle application** (earlier to full throttle). This is a car characteristics story — the RB20 gave Verstappen the mechanical grip to exploit these techniques consistently across all 15 corners of the Bahrain circuit.
